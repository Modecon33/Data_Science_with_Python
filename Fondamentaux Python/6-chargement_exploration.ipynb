### üìå Techniques d'Imputation des Valeurs Manquantes
L‚Äôimputation consiste √† remplacer les valeurs manquantes dans un dataset pour √©viter les erreurs lors de l‚Äôanalyse et du mod√®le de Machine Learning.

### üîπ 1. Suppression des valeurs manquantes
üìå Quand l'utiliser ?

- Si le pourcentage de valeurs manquantes est tr√®s faible (<5%).
- Si la colonne n'est pas essentielle √† l'analyse.

üìå **Code :**  
```python
df.dropna(inplace=True) # Supprime les lignes contenant des valeurs manquantes
dropna(axis=1, inplace=True) # Supprime les colonnes contenant des valeurs manquantes
```
### üîπ 2. Imputation par des valeurs statistiques
üìå Quand l'utiliser ?

- Lorsque la colonne contient des **valeurs num√©riques**.
- Si les valeurs manquantes sont **peu nombreuses**.

üìå M√©thodes courantes :
‚úÖ Moyenne (Mean) : Pour des distributions normales

üìå **Code :**  
```python
df['colonne'].fillna(df['colonne'].mean(), inplace=True)
```

‚úÖ M√©diane (Median) : Si la distribution est asym√©trique (pr√©sence d‚Äôoutliers)

üìå **Code :**  
```python
df['colonne'].fillna(df['colonne'].median(), inplace=True)
```

‚úÖ Mode (valeur la plus fr√©quente) : Pour les variables cat√©goriques

üìå **Code :**  
```python
df['colonne'].fillna(df['colonne'].mode()[0], inplace=True)
```

### üîπ 3. Imputation par une valeur arbitraire
üìå Quand l'utiliser ?

- Si la colonne est cat√©gorielle et contient des valeurs manquantes.
- On peut ajouter une cat√©gorie "Inconnu" ou "Autre".

üìå **Code :**  
```python
df['colonne'].fillna('Inconnu', inplace=True)
```

### üîπ 4. Imputation par la R√©gression ou le k-NN
üìå Quand l'utiliser ?

- Si les valeurs manquantes d√©pendent fortement d‚Äôautres variables.
- Si le dataset est assez grand pour entra√Æner un mod√®le.

üìå M√©thodes :
‚úÖ R√©gression lin√©aire : Pr√©dire les valeurs manquantes √† partir d'autres variables.
‚úÖ k-Nearest Neighbors (k-NN) : Remplacer la valeur manquante par la moyenne des k voisins les plus proches.

üìå **Code :**  
```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
df_imputed = imputer.fit_transform(df)
```
### üîπ 5. Imputation avanc√©e avec Iterative Imputer (MICE)
üìå Quand l'utiliser ?

- Si les donn√©es manquantes d√©pendent de plusieurs variables.
- Si on veut une approche plus robuste qu‚Äôune simple moyenne ou m√©diane.

üìå **Code :**  
```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer()
df_imputed = imputer.fit_transform(df)


| M√©thode                    | Quand l'utiliser ?                     | Exemple                                      |
|----------------------------|---------------------------------------|----------------------------------------------|
| **Suppression**            | Si peu de valeurs manquantes          | `df.dropna()`                               |
| **Moyenne/M√©diane**        | Variables num√©riques                  | `df['col'].fillna(df['col'].mean())`        |
| **Mode**                   | Variables cat√©goriques                | `df['col'].fillna(df['col'].mode()[0])`     |
| **Valeur arbitraire**       | Ajouter une cat√©gorie "Inconnu"       | `df['col'].fillna("Inconnu")`               |
| **k-NN**                   | Bas√© sur les k voisins                | `KNNImputer(n_neighbors=5)`                 |
| **MICE (Iterative Imputer)** | Approche plus robuste                 | `IterativeImputer()`                        |
